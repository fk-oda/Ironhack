## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
#orders.loc[orders.duplicated(keep=False),:]
orders = orders.drop_duplicates()
merged = orders.merge(orders.groupby('CustomerID', as_index=False)[['amount_spent']].sum().rename(columns={'amount_spent':'total_spent'}),
                                                                                                  how='left', on='CustomerID')
merged.head()
```

```python
mask_q90 = merged['total_spent'] >= merged['total_spent'].quantile(0.9)
mask_q75 = merged['total_spent'] >= merged['total_spent'].quantile(0.75)
merged.loc[mask_q90,'Category'] = 'VIP'
merged.loc[mask_q75 & ~mask_q90,'Category'] = 'Preferred'
merged.loc[merged['Category'].isnull(), 'Category'] = 'Regular'
merged.head()
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
unique_customers = merged[['CustomerID', 'Country', 'Category']].drop_duplicates()
unique_customers.loc[unique_customers['Category'] == 'VIP',:].groupby('Country').count().sort_values(by='Category', ascending=False)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
unique_customers_ohe = pd.get_dummies(unique_customers, columns=['Category'], prefix='', prefix_sep='')
unique_customers_ohe['VIP+Preferred'] = unique_customers_ohe['VIP'] + unique_customers_ohe['Preferred']
unique_customers_ohe.groupby('Country').sum().sort_values(by='VIP+Preferred', ascending=False)
```
